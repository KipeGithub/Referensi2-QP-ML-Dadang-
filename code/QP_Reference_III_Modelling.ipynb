**Memanggil data**

```python
import pandas as pd
from google.colab import drive


drive.mount('/content/drive')

df = pd.read_excel('/content/drive/MyDrive/Data_gabung/label.xlsx')

df
```

**TF-IDF dan Splitting dat**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import numpy as np

vectorizer = TfidfVectorizer()
X = df['Review_clean']
y = df['label']

x = vectorizer.fit_transform(X)
y = np.array(y)



x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, test_size=0.3, stratify=y)

feature_names = vectorizer.get_feature_names_out()

df_tfidf = pd.DataFrame(x_test.toarray(), columns=feature_names)
```

```python
print(x_train.shape)
print(x_test.shape)
```

**save data test**

```python
x_tes = pd.DataFrame(df_tfidf)
x_tes.to_csv("/content/drive/MyDrive/Data_gabung/x_test_70.csv", index=False)
y_tes = pd.DataFrame(y_test)
y_tes.to_csv("/content/drive/MyDrive/Data_gabung/y_test_70.csv", index=False)
```

**grid search**

```python
from sklearn import tree
from sklearn.model_selection import GridSearchCV

param_grid = {
    'criterion' : ['gini','entropy'],
	'max_depth' : range(1,15),
	'min_samples_split' : range(1,10),
	'min_samples_leaf' : range(1,5),
}

search = GridSearchCV(tree.DecisionTreeClassifier(), param_grid, scoring='accuracy', cv=10, verbose=3 )
```

```python
result = search.fit(x_train, y_train)
```

**hasil grid search**

```python
print('Best score: %s' % result.best_score_)
print('Best Hyperparameters: %s' % result.best_params_)
```
