```python
#remove duplicates
cleann = pd.DataFrame(cleann)
cleann.drop_duplicates(subset='Review', keep='first', inplace=True)
#cleann = cleann.drop_duplicates(subset=None, keep="first", inplace=True)
#cleann.drop_duplicates(subset="review", keep="first", inplace=True)
print(cleann)
```

**remove simbol**

```python
#removesymbol
import re
import string
def remove_symbol(text):
    text = re.sub(r'[^\w]', ' ', text)
    text = re.sub('[0-9]+', '',  text)
    return (text)
```

```python
cleann['remove_symbol'] = cleann['Review'].apply(remove_symbol).apply(lambda x: remove_symbol(x))
cleann.head(10)
```

**case folding**

```python
def case_folding(text):
  text = text.lower()
  return text
```

```python
cleann['case_folding'] = cleann['remove_symbol'].apply(lambda x: case_folding(x))
cleann.head(5)
```

**remove dan emot**

```python
!pip install emoji
import emoji

def for_emoji(review):
  SMILEY = {":-(":"sad", ":-)":"smiley"}
  word = review.split()
  reformed = [SMILEY[word] if word in SMILEY else word for word in word]
  review = " ".join(reformed)
  return review
```

```python
cleann['remove_emoji'] = cleann['case_folding'].apply(lambda x: case_folding(x))
cleann.head(5)
```

**Tokenization**

```python
import nltk
import string

nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
```

```python
def clean_review(review):
  stopword_indo = stopwords.words('indonesian') # assuming this is defined earlier in the code
  review_tokens = word_tokenize(review) # defining review_tokens by tokenizing review
  review_clean = []

  for word in review_tokens:
    if(word not in stopword_indo and word not in string.punctuation):
      review_clean.append(word)

  return review_clean

cleann['Review_Translate2'] = cleann['Review_Translate2'].astype(str)
cleann['review_clean'] = cleann['Review_Translate2'].apply(lambda x: clean_review(x))

cleann.head()
```

```python
def remove_punct(review):
  review = " ".join([char for char in review if char not in string.punctuation])
  return review

cleann['review_clean_join'] = cleann['review_clean'].apply(lambda x: remove_punct(x))
cleann.head()
```

**stopword**

```python
!pip install pySastrawi
import nltk
from nltk.corpus import stopwords
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
nltk.download('stopwords')
```

```python
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()

stopword_ind = stopwords.words('indonesian')
custom_stopwords = ['yg', 'km', 'rb', 'dgn', 'ig', 'dr', 'ga', 'rp', 'bgt', 'byk', 'utk', 'jgn', 'gtr', 'tp', 'dr', 'dari', 'dgn',
                    'dg', 'g', 'lbh', 'dlm', 'dl', 'sblm', 'spt', 'd', 'nya', 'sdh', 'jln', 'sbg', 'sm', 'aja', 'bgt', 'wkt',
                    'tpi', 'msk', 'n', 'b', 'tpa', 'gt', 'gak', 'gk', 'knp', 'br', 'blm', 'skrg', 't', 'o', 'yah', 'lg',
                    'plg', 'tdk', 'drpd', 'dpt', 'utk', 'dri', 'lgsg', 'nnti', 'jg', 'tp', 'krna', 'tpi', 'bg', 'tuh', 'yo', 'loh',
                    'nah', 'ih', 'ihh', 'duh', 'ad', 'udh', 'lg', 'yahh']
stopword_ind += custom_stopwords

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stopword_ind]
    return ' '.join(filtered_words)

cleann['SAS_removedStopwords'] = cleann['review_clean_join'].apply(remove_stopwords)
print(cleann['SAS_removedStopwords'].head(5))
```

**stemming**

```python
#stemming
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
teks = []
for index, row in cleann.iterrows():
  teks.append(stemmer.stem(row['SAS_removedStopwords']))
cleann['Review_clean']=teks
cleann
```
